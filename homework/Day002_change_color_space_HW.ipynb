使用 [Lena](http://optipng.sourceforge.net/pngtech/img/lena.png)

- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
from cv2 import cv2
import matplotlib.pyplot as plt
```

```python
path = 'lena.png'
img = cv2.imread(path)

HLS = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)

LAB = cv2.cvtColor(img,cv2.COLOR_RGB2LAB)
```

```python
img = img[:,:,::-1]
HLS = HLS[:,:,::-1]
LAB = LAB[:,:,::-1]
```

```python

f, axs = plt.subplots(1,3) 
f.set_figwidth(15)
f.set_figheight(5)
axs[0].imshow(img)
axs[0].set_title('RGB')

axs[1].imshow(HLS)
axs[1].set_title('HLS')

axs[2].imshow(LAB)
axs[2].set_title('LAB')
plt.show()
```
